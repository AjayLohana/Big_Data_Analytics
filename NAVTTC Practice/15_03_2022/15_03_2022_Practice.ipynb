```python
import matplotlib.pyplot as plt
year=[2005,2006,2007,2008]
pop=[5,6,7,8]
plt.plot(year,pop,label='population 1') #plots the data on specified co-ordinates in background
pop2=[6,8,9,10]
plt.plot(year,pop2,label='population 2') #plots the data and assign label to it
plt.xlabel('Independant Variable') #specifies xlabel
plt.ylabel('Dependent variable') #specifies ylabel
plt.title('Interesting Graph')  #specifies title
plt.legend() #visualises labels specified to the data
```

```python
import matplotlib.pyplot as plt
year=[2005,2006,2007,2008]
pop=[5,6,7,8]
plt.plot(year,pop,label='population 1') #plots the data on specified co-ordinates in background
pop2=[6,8,9,10]
plt.plot(year,pop2,label='population 2') #plots the data and assign label to it
pop3=[3,4,6,9]
plt.scatter(year,pop3,label='Population 3')
plt.xlabel('Independant Variable') #specifies xlabel
plt.ylabel('Dependent variable') #specifies ylabel
plt.title('Interesting Graph')  #specifies title
plt.legend() #visualises labels specified to the data
```

```python
import matplotlib.pyplot as plt
year=[2005,2006,2007,2008]
pop=[5,6,7,8]
plt.plot(year,pop,label='population 1',linewidth=5) #plots the data on specified co-ordinates in background
pop2=[6,8,9,10]
plt.plot(year,pop2,label='population 2',linewidth=10) #plots the data and assign label to it
plt.xlabel('Independant Variable') #specifies xlabel
plt.ylabel('Dependent variable') #specifies ylabel
plt.title('Interesting Graph')  #specifies title
plt.legend() #visualises labels specified to the data
```

```python
student=[1,3,5,7]
marks_English=[90,80,95,70]
marks_Urdu=[70,90,85,89]
marks_maths=[85,76,87,90]

plt.scatter(student,marks_English,label='Marks of Students in English',marker='*')
plt.scatter(student,marks_Urdu,label='Maks of student in Urdu',marker='+')
plt.scatter(student,marks_maths,label='Marks of student in Maths')
plt.ylabel('Marks of Students')
plt.xlabel('Students Roll Number')
plt.title('Marks of Students in Different Subjects')
plt.legend()
plt.grid()
```
